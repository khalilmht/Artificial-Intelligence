```python
from sklearn import linear_model
from sklearn.metrics import root_mean_squared_error,r2_score
```

```python
lr=linear_model.LinearRegression()
```

```python

y = df['price']
X=df.drop(columns=['price'])
```

```python
lr.fit(X,y)
```

```python
prix_pred=lr.predict([[4100]])
```

```python
print(prix_pred)
```

```python
plt.xlabel("Area")
plt.ylabel("price")
plt.scatter(df['area'],df['price'],color='red',marker='+')
plt.plot(df['area'],lr.predict(X),color='blue')
```

```python
r2_score(y,lr.predict(X))
```

```python
root_mean_squared_error(y,lr.predict(X))
```

```python
intercept= lr.intercept_
coef = lr.coef_
```

```python
price1 = intercept + coef * 1000
price2 = intercept + coef * 1500
```

```python
print(price1)
print(price2)
```
